```python
from src.data_loader import load_telco_csv
from src.preprocessing import make_xy, clean_telco, one_hot_encode
from src.split import stratified_split
from src.models import make_logistic_regression
from src.evaluation import report, top_coefficients
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score, classification_report
```

# Clean TotalCharges + drop missing

```python
df = load_telco_csv("../../data/raw/WA_Fn-UseC_-Telco-Customer-Churn.csv")
X, y = make_xy(df)
X, y = clean_telco(X, y)
```

# Now, we gonna one-hot encode all the other categorical object because there are not numerical

```python
X = one_hot_encode(X)
```

# Now, let's split Train/test dataset

```python
X_train, X_test, y_train, y_test = stratified_split(X, y)
```

# Baseline model , so this baseline model, will ignores all features and always predict the most common class:

### 1- We want to compare our model to this baselline cause if the baseline did better than our model, then there is no point using ML there cause it's not really learning or the features aren't that good

```python
baseline = DummyClassifier(strategy="most_frequent")
baseline.fit(X_train, y_train)

y_pred_baseline = baseline.predict(X_test)
print("Baseline accuracy:", accuracy_score(y_test, y_pred_baseline))
print(classification_report(y_test, y_pred_baseline))
```
